# **Product and Gender Analysis**

```python
print(df.columns)
```

```python
s = df.groupby(["Product", "Gender"]).size().reset_index(name="Count")
print(s)
```

```python
pivot_df = s.pivot(index="Product", columns="Gender", values="Count")

# Plot
pivot_df.plot(kind="bar", stacked=True)
plt.title("Number of Products Used by Gender (Stacked)")
plt.xlabel("Product")
plt.ylabel("Count")
plt.show()
```

**Observations**


*   In the male Category the most used product is TM195
*   In TM798 female users are in majority than the male category unlike other products

# **Correlation**
Let's plot multivariate charts between variables to understand their interaction with each other.

```python
# Check for correlation among numerical variables
num_var = ['Age', 'Education' , 'Usage', 'Fitness', 'Income', 'Miles']
corr = df[num_var].corr()

# Plot the heatmap
plt.figure(figsize = (14, 10))
sns.heatmap(corr, annot = True, cmap = 'coolwarm',
        fmt = ".1f",
        xticklabels = corr.columns,
        yticklabels = corr.columns)
```

**Observations**
1. In this Usage and Miles have the correlation 0.8 which is almost 1.
2. Fitness and Usage has the correlation of of 0.7 which is indicates that the fitness is depenedent on the usage of the product.

```python
grouped_totals = df.groupby(["Product"]) ["Miles"].mean()
print(grouped_totals)
```
